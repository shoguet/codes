You will see that when you increase the number of people and bins the histogram will start to look like a bell curve. This is because I am extracting the numbers for the age from something call the normal distribution (or a guaussian distribution). Below I have the plot of the normal distribution or gaussian distribution. (10,000 people and 20 bins). 

Equation for a guassian distribution is,

<font size = '5'>$$ f(x) = \frac{1}{\sqrt{2\pi{\sigma}^2}}{e}^{\frac{-1}{2}{\frac{(x - \mu)^2}{\sigma^2}}}$$ </font>

In the equation, 
'x' is the random variable. 
$\mu$ is the mean of the random variable.(in our case it is the age) 
$\sigma^2$ is the variance of the random variable. (Variance is the measurement that calculates how far away each variable is from the mean) 

Also the gaussian distribution gives the probability density. (The probability of a certain age for example). I have created a sample gaussian distribution below. We can talk about this next week. You can see that when the number of people increases the histogram looks like a gaussian distribution. However the histogram gives the number of people in that certain bin and the gaussian distribution gives how probable a person will fall in to a specific age. (for example if you look at the plot below, out of the 10,000 people if we randomly pick a person the probability of that person will be 50 years old is about 0.12)

```python
#this is the function for the guassian distribution
def my_gaussian(sigma, mu,x):
    
    gaussian = 1/(np.sqrt(sigma*2*np.pi))*np.exp(-1/2*(x - mu)**2/sigma)
    #gaussian = gaussian/np.sum(gaussian)
    
    return gaussian

age = []
nofpeople = 10000

for i in range(nofpeople):
    
    #import a random number between 0 and 100
    a = np.random.normal(loc = 50, scale = 10)
    age.append(a)


age_gaussian = np.arange(np.min(age), np.max(age), 0.1)
gaussian = my_gaussian(10, 50, age_gaussian)


plt.title('Gaussian Distribution')
plt.plot(age_gaussian, gaussian)
plt.xlabel('Age')
plt.ylabel('Probability Density')
plt.show()
```

```python
#gaussian and cumulative gaussian
import ipywidgets as widgets 
def my_gaussian(mu = 0, sigma = 10):
    
    #x values
    x = np.arange(-10, 10, 0.1)
    
    #gaussian equaton
    gaussian = 1/(np.sqrt(sigma*2*np.pi))*np.exp(-1/2*(x - mu)**2/sigma)
    gaussian = gaussian/np.sum(gaussian)
    
    #cumulative gaussian
    cum_gaussian = np.zeros(len(gaussian))
    cum_gaussian[0] = gaussian[0]
    
    for i in range(len(gaussian)-1):

        cum_gaussian[i+1] = cum_gaussian[i] + gaussian[i+1]   
    
     
    #plot the two graphs
    plt.xkcd()
    
    fig, axs = plt.subplots(2)
    fig.set_figheight(8)
    fig.set_figwidth(6)
    axs[0].plot(x, gaussian)
    axs[1].plot(x, cum_gaussian)

_ = widgets.interact(my_gaussian, mu = (-10,10, 0.5), sigma = (1, 10, 0.5))
```
