```python
#b.
import numpy as np
import matplotlib.pyplot as plt
def first_gaussian(sigma, mu, x):
   
    gaussian= 1/(np.sqrt(sigma*2*np.pi))*np.exp(-1/2*(x-mu)**2/sigma)
    #gaussian = gaussian/np.sum(gaussian)
    return gaussian

stimulus = []
probability = 1000

for i in range(probability):
    a = np.random.normal(loc = 4, scale = 10)
    stimulus.append(a)
    
stimulus_gaussian = np.arange(np.min(stimulus), np.max(stimulus), 0.1)
gaussian = first_gaussian(10,4,stimulus_gaussian)
    
plt.title("figure 6")
plt.ylabel("probability")
plt.xlabel("hypothesized stimulus s")
plt.plot(stimulus_gaussian, gaussian, color = 'm', label = "posterior, p(s|xtrial)")

gaussian = first_gaussian(10,8,stimulus_gaussian)
plt.plot(stimulus_gaussian, gaussian, label = "likelihood, L(s)=p(Xtrial|s)")

gaussian = first_gaussian(100,-2, stimulus_gaussian)
plt.plot(stimulus_gaussian, gaussian, color = 'r', label = "prior, p(s)")

plt.legend()
plt.show()
```
